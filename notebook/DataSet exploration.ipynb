```python
import matplotlib.pyplot as plt
import pandas as pd
```

# Show first five rows of the data set

```python
df = pd.read_csv('../data/dataset.csv')
print(df.head())
```

# Show how many diseases in data set and their count

```python
plt.figure(figsize=(12, 8))
classes_proportion = df['Disease'].value_counts().reset_index()
plt.bar(range(len(classes_proportion)), classes_proportion.iloc[:, 1])
plt.xticks(range(len(classes_proportion)), classes_proportion.iloc[:, 0], rotation=90)
plt.show()
```
